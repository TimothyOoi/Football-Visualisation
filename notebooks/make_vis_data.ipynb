## 2. First Choice Goalkeeper Dataset (2021/2022 season)

```python
team_standings_df = pd.read_csv('../vis_data/team standings 2021-2022.csv')
```

```python
team_standings_df.head()
```

```python
team_standings_df.shape
```

```python
keepers_df = pd.read_csv('../datasets/Big5/2021-2022/keepers.csv')
keepers_df.head()
```

```python
keepers_adv_df = pd.read_csv('../datasets/Big5/2021-2022/keepersadv.csv')
keepers_adv_df.head()
```

```python
# merge team standings with standard goalkeeper stats dataset on goalkeeper and squad
temp_df = team_standings_df.merge(keepers_df, left_on=["Squad", "Goalkeeper"], right_on=["Squad", "Player"])
```

```python
final_df = temp_df.merge(keepers_adv_df, left_on=["Squad", "Goalkeeper"], right_on=["Squad", "Player"])
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df = final_df.drop(columns=['Country', 'Player_y', 'Nation_y', 'Pos_y', 'Comp_y', 'Age_y', 'Born_y', '90s', 'Goals GA', 'Performance W', 'Performance D', 'Performance L'])
```

```python
final_df = final_df.rename(columns={'Player_x': 'Player', 'Nation_x':'Nation', 'Pos_x': 'Pos', 'Comp_x': 'Comp', 'Age_x': 'Age', 'Born_x': 'Born'})
```

```python
final_df.to_csv('../vis_data/gk_and_team_standings.csv', index=False)
```
